```python
import matplotlib.pyplot as plt
names = ["Mans", "Amina", "Chris", "Chorni", "Sanzh"]
scores = [100, 88, 92, 52, 67]

plt.plot(names, scores, marker='o')
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()

plt.bar(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()

```
