```python
import numpy as np
arr1 = np.arange(1, 11)
print("1D Array:", arr1)
arr2 = np.random.randint(1, 20, size=(3, 3))
print("2D Array:\n", arr2)
print("Sum:", arr1.sum())
print("Mean:", arr1.mean())
print("Max:", arr1.max())

arr_multiplied = arr1 * 2
print("Multiplied by 2:", arr_multiplied)


arr_replaced = arr1.copy()
arr_replaced[arr_replaced>5] = 0
print("Values > 5 replaced with 0:", arr_replaced)
```
